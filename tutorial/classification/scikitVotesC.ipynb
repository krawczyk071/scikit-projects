```python
# Import pandas 
import pandas as pd
import numpy as np
  
# reading csv file
# r stands for raw string, helps with unicode error (slashes)
filepath = r"C:\Users\krawc\OneDrive\Documents\DB\datasets\datacamp-master\Supervised Learning with scikit-learn\data\housevotes.csv"
#df1 = pd.read_csv(filepath, encoding = 'utf8', index_col=0)
df = pd.read_csv(filepath, encoding = 'utf8')
```

```python
df.head()
```

```python
df[df == 'y'] = 1
df[df == 'n'] = 0
```

```python
mapping={'democrat':1,'republican':0}
df['party']=df.party.map(mapping)
```

```python
# Convert '?' to NaN
df[df == '?'] = np.nan

# Print the number of NaNs
print(df.isnull().sum())

```

```python
y = df.party.values
y = y.reshape(-1, 1)
X = df.drop('party' ,axis='columns')
```

```python
# Import the Imputer module
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

# Setup the Imputation transformer: imp
imp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')

# Instantiate the SVC classifier: clf
clf = SVC()

# Setup the pipeline with the required steps: steps
steps = [('imputation', imp),
        ('SVM', clf)]
```

```python
from sklearn.metrics import confusion_matrix, classification_report

# Create the pipeline: pipeline
pipeline = Pipeline(steps)

# Create training and test sets
X_train, X_test, y_train, y_test = train_test_split(X , y, test_size=0.3 , random_state=42)

# Fit the pipeline to the train set
pipeline.fit(X_train,y_train)

# Predict the labels of the test set
y_pred = pipeline.predict(X_test)

# Compute metrics
print(classification_report(y_test,y_pred))
```
